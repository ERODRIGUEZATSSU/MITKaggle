# Movie Keywords

```python
movie_keywords = pd.read_csv("movies_keywords.csv.zip", compression='zip')
```

```python
movie_keywords.head()
```

```python
movie_keywords.shape
```

```python
movie_keywords.info()
```

```python
#turn object to a list of all the values with key="name"
for i in range(len(movie_keywords.keywords)):
    keywords_list = []
    keywords = eval(movie_keywords.keywords[i])
    for key, value in [(key, value) for x in keywords for (key, value) in x.items()]:
        if key == "name":
            keywords_list.append(value)
#     print(genres_list)
    movie_keywords.at[i, 'new_keywords'] = (keywords_list)
```

```python
movie_keywords.new_keywords
```

```python
movie_keywords = movie_keywords.drop(['keywords'], axis=1)
movie_keywords.head()
```

```python
type(movie_keywords.new_keywords)
```

#### Multilinear Binerizer

```python
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
```

```python
movie_keywords_mlb = pd.DataFrame(mlb.fit_transform(movie_keywords.new_keywords), columns=mlb.classes_)
```

```python
movie_keywords_mlb.head()
```

```python
movie_keywords_mlb.shape
```

### Merge movie_keywords df and movie_keywords_mlb df

```python
movie_keywords_merged = pd.concat([movie_keywords, movie_keywords_mlb], axis=1, join='inner')
```

```python
movie_keywords_merged.head()
```

```python
#drop "new_keywords" column 
movie_keywords_merged = movie_keywords_merged.drop(['new_keywords'], axis=1)

movie_keywords_merged.head()
```

```python
compression_opts = dict(method='zip',
                        archive_name='final_keywords.csv')  

movie_keywords_merged.to_csv('final_keywords.zip', 
                             compression=compression_opts)  
```

# Movie Metadata 

```python
movies_metadata_prepro = pd.read_csv("movie_metadata_preprocessed.zip", compression='zip')
```

```python
movies_metadata_prepro.head()
```

### One hot encoding

```python
one_hot_status = pd.get_dummies(movies_metadata_prepro['status'], prefix="status_")
# adult one hot encoded
one_hot_adult = pd.get_dummies(movies_metadata_prepro['adult'], prefix="adult_")
# video one hot encoded
one_hot_video = pd.get_dummies(movies_metadata_prepro['video'], prefix="video_")
# imbdb one hot encoded
one_hot_imbdb = pd.get_dummies(movies_metadata_prepro['imdb_id'], prefix="imdb_id_")
# original_language one hot encoded
one_hot_original_language = pd.get_dummies(movies_metadata_prepro['original_language'], prefix="original_language_")
# original_title one hot encoded
one_hot_original_title = pd.get_dummies(movies_metadata_prepro['original_title'], prefix="original_title_")
# title one hot encoded
one_hot_title = pd.get_dummies(movies_metadata_prepro['title'], prefix="title_")

#join
one_hot_merged = pd.concat([one_hot_status, one_hot_adult, one_hot_video, one_hot_imbdb, one_hot_original_language, one_hot_original_title, one_hot_title], axis=1, join='inner')
one_hot_merged
```

```python
#drop poster_path and overview columns
movies_metadata_prepro = movies_metadata_prepro.drop(['overview', 'title','poster_path','status', 'adult', 'video', 'imdb_id', 'original_language', 'original_title'], axis=1)
```

### Apply PCA to reduce the dimentionality of the data

```python
# first find the number of components using MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_rescaled = scaler.fit_transform(df)
```

```python
pca = PCA().fit(df_rescaled)
```

```python
% matplotlib inline
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (12,6)
fig, ax = plt.subplots()
xi = np.arange(1, 11, step=1)
y = np.cumsum(pca.explained_variance_ratio_)

plt.ylim(0.0,1.1)
plt.plot(xi, y, marker='o', linestyle='--', color='b')

plt.xlabel('Number of Components')
plt.xticks(np.arange(0, 11, step=1)) #change from 0-based array index to 1-based human-readable label
plt.ylabel('Cumulative variance (%)')
plt.title('The number of components needed to explain variance')

plt.axhline(y=0.95, color='r', linestyle='-')
plt.text(0.5, 0.85, '95% cut-off threshold', color = 'red', fontsize=16)

ax.grid(axis='x')
plt.show()
```

```python
# 95% or 99% of variance 
from sklearn.decomposition import PCA
pca = PCA(n_components = 0.95)
pca.fit(df_rescaled)
reduced = pca.transform(df_rescaled)
```
